### Pedagogical Style

The format for the course is 

* Theory 

* Practise

* Theory

* Repeat

The idea behind this is that a little information up front helps get the processes started (first Theory, ie *motivation*). But you never learn anything without experimenting a lot on your own (Practise). After you have learned basically how to do something well enough on your own, it helps a lot to synthesize the information in a coherent story (second Theory, ie deeper understanding).

# Course Description

This course will teach the basic principles of modern programming and computational analysis. We will use the programming language **Python** for all of our calculations. Python provides a very straightforward and efficient way to learn and implement basics programming concepts. It also has the benefit of a giant user community and a lot of great online resources for learning and troubleshooting. 
We will learn to use in a practical way the basic ingredients of *any* programming language, i.e.,

    Syntax e.g.: ( ), versus [ ] , versus { }   .

    Variables e.g.: float, complex, integer, string.

    Control structures e.g.: for loops, while loops, if statements, functions.

    Data structures e.g.: lists, arrays, dictionaries, classes, environments.

    Tools e.g.: notebooks, visualisation, external libraries, etc.
    
There are a small handful of other similar principles that we will discuss occasionally; e.g.: memory management, optimisation and parallelism. But these are slightly more technical and not necessary to make a lot of progress.  Python makes learning and applying easy and quick without needing to fuss with too much tedious book keeping (e.g. FORTRAN, C/C++). It's more established than newer options (e.g. Julia). It's free and open source (unlike e.g. Matlab).

Using these concepts in Python, we will explore the fundamentals of computational/numerical analysis. We will learn the basics ingredients needed to solve *any* practical scientific computing problem.  Some key examples are:

    How to represent floating-point decimal numbers on a computer. 

    The uses and dangers of iteration and recursion.

    Sources and control of errors in computations. 

    Leveraging and computing linear operations on data.

These four principles form the foundation of virtually all mathematical computing problems. We will see the different ways to apply the basic principles in a series of topics and applications.

The course will cover an overview of several different aspects of mathematical computing. These topics will sometimes be independent from each other. And sometimes they will have a lot of ideas in common. Many times, we will focus heavily on a topic not because that topic is extremely important all on its own. We will talk deeply about most topics *because they teach you important ideas about how to think about and solve difficult problems.* Expert problem solving is one of the most important and difficult skills in life. If you need to use computers to solve real-world problems, there are many excellent software libraries written by world experts.*And you should use these libraries*. This course will teach you a lot about how things work, when you should use an existing tool, and when you might need to make your own.
